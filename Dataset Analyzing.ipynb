```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
import re
```

```python
page=requests.get("https://www.makaan.com/hyderabad-residential-property/buy-property-in-hyderabad-city?page=2")
page1=BeautifulSoup(page.text)

p9=page1.find_all("a",class_="projName")
building_Name=[]
for i in p9:
    building_Name.append(i.text)
    
    
p=page1.find_all("div",class_="seller-info")
Constructed_by=[]
for i in p:
    Constructed_by.append(i.text)
    
    
p1=page1.find_all("a",class_="loclink")
Location=[]
for i in p1:
    Location.append(i.text)
    
    
p2=page1.find_all("td",class_="lbl rate")
Prices_per_sqft=[]
for i in p2:
    a=i.text
    Prices_per_sqft.append(re.findall(r"(\d+\W\d+)",a)[0])
    
    
p3=page1.find_all("td",class_="price")
Price=[]
for i in p3:
    Price.append(i.text)
    
    
p4=page1.find_all("td",class_="size")
Area_sq_ft=[]
for i in p4:
    Area_sq_ft.append(i.text)
    
    
p5=page1.find_all("td",class_="val")
Building_Status=[]
for i in p5:
    Building_Status.append(i.text)
    
    
p6=page1.find_all("li",class_="keypoint")
Possession_on=[]
for i in p6:
    a=i.text
    b=re.findall(r"\w+\s\w+(\s\w+\s\d+)",a)
    if len(b)>0:
        Possession_on.append(b[0])
        
        
p7=page1.find_all("div",class_="title-line-wrap")
House_size=[]
for i in p7:
    a=i.text
    b=re.findall(r"\d+\s\w+",a)
    if len(b)>0:
        House_size.append(b[0])
    else:
        House_size.append(np.nan)
        
        
p8=page1.find_all("div",class_="title-line")
Building_type=[]
for i in p8:
    a=i.text
    Building_type.append(re.findall(r"\w+",a)[2])
```
